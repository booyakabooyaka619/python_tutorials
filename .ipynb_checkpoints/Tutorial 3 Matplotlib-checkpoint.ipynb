```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(10,0,-1)
```

```python
x
```

```python
y
```

```python
# scatter plot using matplotlib
plt.scatter(x,y,c="r")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Example of scatter plot")
```

```python
# scatter plot with straight line
plt.plot(x,y,"r*--")
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
